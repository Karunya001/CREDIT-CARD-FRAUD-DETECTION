```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_card_data=pd.read_csv('creditcard.csv')
```

```python
credit_card_data['Class'].value_counts()
```

```python
approve=credit_card_data[credit_card_data.Class== 0]
print(approve.shape)
```

```python
fraud=credit_card_data[credit_card_data.Class== 1]
print(fraud.shape)
```

```python
approve.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
credit_card_data.groupby('Class').mean()
```

```python
#creating new sample data from the give data set
approve_sample=approve.sample(n=492)
```

```python
new_dataset = pd.concat([approve_sample , fraud],axis =0 )
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
x=new_dataset.drop(columns='Class' , axis =1)
print(x)
```

```python
y=new_dataset['Class']
print(y)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,stratify=y,random_state=2)
```

```python
print(x.shape)
print(x_train.shape)
print(x_test.shape)
print(y.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
pred_train=model.predict(x_train)
acc_score=accuracy_score(pred_train,y_train)
```

```python
print("ACCURACY:\n",acc_score)
```

```python
pred_test=model.predict(x_test)
acc_score_test=accuracy_score(pred_test,y_test)
```

```python
print("ACCURACY:\n",acc_score_test)
```
